1. Flask is a micro web framework written in Python. It has no database abstraction layer, form validation, or any other components where pre-existing third-party libraries provide common functions.

Advantages of Flask framework:
1) It is a lightweight framework that is easy to learn.
2) Provide flexibility to the developer to experiment with their modules or architecture.
3) Easily scalable for the applications.

```python
## 2.
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello World!! </h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

3. App routing is the technique used to map the specific URL with the associated function intended to perform some task. 

To bind a function to an URL path we use the app.route decorator.

```python
## 4.
from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "<h1> Welcome to ABC Corporation </h1>"
# https://orange-accountant-odetm.pwskills.app:5000/welcome
   
@app.route("/")
def welcome1():
    return """Company Name: ABC Company <br>
            Location: India  <br>
            Contact detail: 999-999-999""" 
# https://orange-accountant-odetm.pwskills.app:5000

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
